```python
import pandas as pd
```

```python
subjects = pd.read_excel("subjects.xlsx").set_index('Unnamed: 0')
subjects
```

```python
columns = subjects.iloc(0)
# subjects[1:]
columns[0]
```

```python
def prefix_adder(key, text):
    prefix = "Anonymized " if (key != 'Anonymized' and key != 'Raw' and int(key[9:]) <= 3) or (key == 'Anonymized') else "Raw "
    return prefix + text

new_columns = {}

for column in columns[0].keys():
    new_columns[column] = prefix_adder(column, columns[0][column])

subjects.rename(columns=new_columns, inplace=True)
```

```python
# subjects = subjects[0:]
for column in subjects.columns:
    column_type = type(subjects[column][0])
    if column_type == 'float':
        subjects[column].fillna(0.0, inplace=True)
    elif column_type == 'int':
        subjects[column].fillna(0, inplace=True)
    elif column_type == 'str':
        subjects[column].fillna('', inplace=True)
```
